```python
# Function to convert to Strings  
def listToString(s):  
    
    # initialize an empty string 
    str1 = ""  
    
    # traverse in the string   
    for ele in s:  
        str1 += ele   
    
    # return string   
    return str1  
```

```python
#function to find all floats values from list
def solve(lis):
    for x in lis:
        try:
            float(x)
            return True
        except:
            return False
```

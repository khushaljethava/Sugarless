```python
# Function to convert to Strings  
def listToString(s):  
    
    # initialize an empty string 
    str1 = ""  
    
    # traverse in the string   
    for ele in s:  
        str1 += ele   
    
    # return string   
    return str1  
```

```python
#function to find all floats values from list
def solve(lis):
    for x in lis:
        try:
            float(x)
            return True
        except:
            return False
```

```python
# OCR Works Here

try:
    from PIL import Image
except ImportError:
    import Image
import pytesseract

def ocr_core(filename):
  
    text = pytesseract.pytesseract.image_to_string(Image.open(filename))
    return text  # Then we will print the text in the image
```

```python

# Function for preparing images from pdf
def prepare_images(pdf_path):
 
    with(Image(filename=pdf_path, resolution=300, width=600)) as source:
        images = source.sequence
        pages = len(images)
        for i in range(pages):
            
            Image(images[i]).save(filename='image' + str(i) + '.png')
            
```

```python
# Main Functions for OCR
images = convert_from_path('test_new.pdf')
img=[]
texts = []
for i, image in enumerate(images):
    fname = "image" + str(i) + ".png"

    image.save(fname, "PNG")
    text  =  ocr_core(fname)
    #os.remove(fname)
    text = listToString(text)  
    texts.append(text)
    text1 = listToString(texts)
    os.remove(fname)
 
```
